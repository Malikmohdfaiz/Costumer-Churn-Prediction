```python
import pandas as pd
import numpy as np
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import recall_score
from sklearn.metrics import confusion_matrix
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier


from imblearn.combine import SMOTEENN
```

```python
dataset=pd.read_csv("cleaned.csv")
```

```python
dataset.head()
```

```python
dataset=dataset.drop('Unnamed: 0',axis=1)
```

```python
x=dataset.drop('Churn',axis=1)
x
```

```python
y=dataset['Churn']
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

# Decision Tree Classifier

```python
model_dt=DecisionTreeClassifier(criterion = "gini",random_state = 100,max_depth=10)
```

```python
model_dt.fit(x_train,y_train)
```

```python
y_pred=model_dt.predict(x_test)
y_pred
```

```python
model_dt.score(x_test,y_test)
```

```python
print(classification_report(y_test,y_pred,labels=[0,1]))
```

```python
print(np.unique(y, return_counts=True))
```

```python
sm = SMOTEENN()

# Resample the dataset
X_resampled, y_resampled = sm.fit_resample(x, y)
```

```python
x1_train,x1_test,y1_train,y1_test=train_test_split(X_resampled, y_resampled,test_size=0.2)
```

```python
model_rf_smote=RandomForestClassifier(n_estimators=100,criterion='gini',random_state=100)
```

```python
model_rf_smote.fit(x1_train,y1_train)
```

```python
yr_predict = model_rf_smote.predict(x1_test)
yr_predict
```

```python
model_score= model_rf_smote.score(x1_test,y1_test)
model_score
```
